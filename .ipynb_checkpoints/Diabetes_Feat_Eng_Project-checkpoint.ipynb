Projektbeschreibung: 

Vorhersage von Diabetes mithilfe von Machine Learning
Ziel: Ein Machine-Learning-Modell soll entwickelt werden, das anhand gegebener Merkmale vorhersagen kann, ob eine Person an Diabetes erkrankt ist. Vor der Modellierung werden umfassende explorative Datenanalysen (EDA) sowie Schritte der Merkmalsgenerierung (Feature Engineering) durchgeführt.

Datensatzinformationen:
Quelle: National Institute of Diabetes and Digestive and Kidney Diseases (NIDDK), USA

Population: Frauen mit pima-indigenem Hintergrund im Alter von 21 Jahren und älter, wohnhaft in Phoenix, Arizona

Größe: 768 Beobachtungen, 8 numerische erklärende Variablen

Zielvariable: Outcome

1 → positives Diabetes-Testergebnis

0 → negatives Diabetes-Testergebnis

| **Variable**                | **Beschreibung**               | **Erklärung**                                                                         |
|-----------------------------|--------------------------------|---------------------------------------------------------------------------------------|
| **Pregnancies**             | Anzahl der Schwangerschaften   |0-6 (Normal), 7<= (hohes Risiko)                                                       |
| **Glucose**                 | Blutzuckerspiegel              |<70 (Niedrig), 70-99 (Normal), 100-125 (Prediabet), 126+Diyabet                        |
| **BloodPressure**           | Blutdruck (diastolisch) (mmHg) |<60 (Niedrig), 60-80 (Normal), >80 (Bluthochdruck)                                     |
| **SkinThickness**           | Hautfalten-Dicke (mm)          |<10 (Niedrig), 10-40 (Normal), >40 (Dicke Haut)                                        |
| **Insulin**                 | Serum-Insulinmessung (μU/ml)   |<16 (Niedrig), 16-166 (Normal), >166 (Diabetes_Risiko)                                 |
| **BMI**                     | Body-Mass-Index ((kg)/(m²))    |<18.5 (Niedrig), 18.5-24.9 (Normal), 25-29.9 (Übergewicht), 30+(Adipos)                |
| **DiabetesPedigreeFunction**| Erblicher Diabetesrisikofaktor | **Hoher DPF** → starke genetische Veranlagung                                         |
|                             |                                | **Niedriger DPF** → schwache genetische Verbindung                                    |
| **Age**                     | Alter (Jahre)                  | 15-25 (Junge Erwachsene), 26-35 (Frühes Erwachsenalter),                              |
|                             |                                | 36-45 (Mittleres Erwachsenalter), 46-59 (RaifesErwachsenalter), 60+ (Senioren)        |
| **Outcome**                 | Diabetesdiagnose               | **1** → positives Testergebnis                                                        |
|                             |                                | **0** → negatives Testergebnis                                                        |

    #IQR-Boxplot

```python
def thresholds_outlier(dataframe, col_name, q1=0.25, q3=0.75):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
```

```python
def outliers_grab(dataframe, col_name, index=False):
    low, up = thresholds_outlier(dataframe, col_name)

    outliers = dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)]
    outlier_count = outliers.shape[0] 

    print(f"Anzahl der Ausreißer ({col_name}): {outlier_count}")

    if outlier_count > 0:
        if outlier_count > 10:
            print(outliers.head())
        else:
            print(outliers)

    if index:
        return outliers.index
```

```python
outlier_indices = []

for col in dia.columns:
    print(f"\nAusreißer Analyse: {col}")
    idx = outliers_grab(dia, col, index=True)
    if idx is not None:
        outlier_indices.extend(idx)
```

```python
def outlier_columns(df, q1=0.25, q3=0.75):
    cols_with_outliers = []
    numeric_cols = dia.select_dtypes(include=["int64", "float64"]).columns.drop(["Outcome", "Age"])

    for col in numeric_cols:
        Q1 = df[col].quantile(q1)
        Q3 = df[col].quantile(q3)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        if df[(df[col] < lower) | (df[col] > upper)].shape[0] > 0:
            cols_with_outliers.append(col)

    return cols_with_outliers
```

```python
outlier_cols = outlier_columns(dia)
print("Spalten mit Ausreißenwerte:", outlier_cols)
```

```python
total_outlier_rows = len(set(outlier_indices))
print(f" Nach dem IQR wurden in {total_outlier_rows} Beobachtungen Ausreißer identifiziert.")
```

    #Umgang mit Ausreißerwerten

1- Ausreißern löschen

```python
def thresholds_outlier_all(dataframe, q1=0.25, q3=0.75):
    limits = {}
    numeric_cols = dataframe.select_dtypes(include=["float64", "int64"]).columns.drop(["Outcome", "Age"])
    for col in numeric_cols:
        low, up = thresholds_outlier(dataframe, col, q1, q3)
        limits[col] = (low, up)
    return limits
```

```python
def remove_outliers(dia, outlier_limits):
    dia_clean = dia.copy()
    for col, (low, up) in outlier_limits.items():
        dia_clean = dia_clean[(dia_clean[col] >= low) & (dia_clean[col] <= up)]
    return dia_clean

outlier_limits = thresholds_outlier_all(dia)
dia_clean = remove_outliers(dia, outlier_limits)
```

2- Winsorizing (re-assignment with tresholds):

```python
numeric_cols = dia.select_dtypes(include=["float64", "int64"]).columns.drop(["Outcome", "Age"])

plt.figure(figsize=(9, 5))
sns.boxplot(data=dia[numeric_cols], orient="h", palette="pastel")
plt.title("Ausreißerverteilung")
plt.tight_layout()
plt.show()
```

```python
def winsorize_dataframe(dia, outlier_limits):
    dia_winsorized = dia.copy()
    
    for col, (low, up) in outlier_limits.items():
        dia_winsorized[col] = dia_winsorized[col].clip(lower=low, upper=up)
    
    return dia_winsorized
# 1. Grenzwerte für Ausreißer berechnen
outlier_limits = thresholds_outlier_all(dia)

# 2. Erstelle den winsorisierten Datensatz
dia_winsor = winsorize_dataframe(dia, outlier_limits)
```

```python
numeric_cols = dia_winsor.select_dtypes(include=["float64", "int64"]).columns.drop(["Outcome", "Age"])

plt.figure(figsize=(9,5))
sns.boxplot(data=dia_winsor[numeric_cols], orient="h", palette="pastel")
plt.title("Ausreißerverteilung (Winsorized)")
plt.tight_layout()
plt.show()
```

    # Binning

```python
dia.loc[dia['Pregnancies'] <= 6, 'Preg_Cat'] = '0-6'
dia.loc[dia['Pregnancies'] >= 7, 'Preg_Cat'] = '7+'

dia['Preg_Cat'].value_counts()
```

```python
dia.loc[dia['Glucose'] < 100, 'Glu_Cat'] = 'Normal' 
dia.loc[(dia['Glucose'] >= 100) & (dia['Glucose'] <= 125), 'Glu_Cat'] = 'Prediabet'
dia.loc[dia['Glucose'] > 125, 'Glu_Cat'] = 'Diabet'

dia['Glu_Cat'].value_counts()
```

```python
dia.loc[dia['BloodPressure'] < 60, 'BP_Cat'] = 'Niedrig' 
dia.loc[(dia['BloodPressure'] >= 60) & (dia['BloodPressure'] < 80), 'BP_Cat'] = 'Normal'
dia.loc[dia['BloodPressure'] >= 80 , 'BP_Cat'] = 'Hoch'

dia['BP_Cat'].value_counts()
```

```python
dia.loc[dia['SkinThickness'] < 40, 'ST_Cat'] = 'Normal' 
dia.loc[dia['SkinThickness'] >= 40 , 'ST_Cat'] = 'Dick'

dia['ST_Cat'].value_counts()
```

```python
dia.loc[dia['Insulin'] < 166, 'Ins_Cat'] = 'Normal' 
dia.loc[dia['Insulin'] >= 166 , 'Ins_Cat'] = 'DiabetRisiko'

dia['Ins_Cat'].value_counts()
```

```python
dia.loc[dia['BMI'] < 25, 'BMI_Cat'] = 'Normal'
dia.loc[(dia['BMI'] >= 25) & (dia['BMI'] <= 29.9), 'BMI_Cat'] = 'ÜberGewicht'
dia.loc[dia['BMI'] > 29.9, 'BMI_Cat'] = 'Adipos'

dia['BMI_Cat'].value_counts()
```

```python
dia.loc[dia['Age'] <= 25, 'Age_Cat'] = 'Junge Erwachsene'
dia.loc[(dia['Age'] >= 26) & (dia['Age'] <= 35), 'Age_Cat'] = 'Frühes Erwachsenenalter'
dia.loc[(dia['Age'] >= 36) & (dia['Age'] <= 45), 'Age_Cat'] = 'Mittleres Erwachsenenalter'
dia.loc[(dia['Age'] >= 46) & (dia['Age'] <= 59), 'Age_Cat'] = 'Reifes Alter'
dia.loc[dia['Age'] >= 46, 'Age_Cat'] = 'Senior'

dia['Age_Cat'].value_counts()
```

```python
dia= dia.copy()
```

```python
cat_columns = [col for col in dia.columns if '_Cat' in col]
dia_encoded = pd.get_dummies(dia, columns=cat_columns, drop_first=True)
dia_encoded = dia_encoded.astype(int)
dia_encoded.head()
```

```python
X = dia_encoded.drop(['Outcome', 'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Age', 'DiabetesPedigreeFunction'], axis=1)
y = dia_encoded['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
# Modal Beschreibung
rf_model = RandomForestClassifier(n_estimators=100, random_state=17)

rf_model.fit(X_train, y_train)

# Prediction
y_pred = rf_model.predict(X_test)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Detalierte Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
def importance_ploting(model, features, num=len(X), save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",
                                                                      ascending=False)[0:num])
    plt.title('Features')
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig('importances.png')


importance_ploting(rf_model, X_train)
```

    # Feature Creation

```python
dia_encoded['Risk_Adipos_Insulin_High'] = (
    (dia['BMI_Cat'] == 'Adipos') & 
    (dia['Insulin'] >= 166)
).astype(int)
```

```python
dia_encoded['Risk_Null_SK_Jung'] = (
    (dia['ST_Cat'] == 'Normal') & 
    (dia['Age_Cat'] == 'Junge Erwachsene')
).astype(int)
```

```python
dia_encoded['Risk_Null_SK_Jung'] = (
    (dia['Glucose'] <= 124) & 
    (dia['BP_Cat'] == 'Niedrig')
).astype(int)
```

```python
dia_encoded['Risiko'] = (
    (dia['BMI_Cat'] == 'Adipos') & 
    (dia['Insulin'] >= 166) & (dia['Glucose'] >= 125) &  (dia['BP_Cat'] == 'Hoch')
).astype(int)
```

```python
dia_encoded['Risiko_Preg'] = (
    (dia['BMI_Cat'] == 'ÜberGewicht') & (dia['Preg_Cat'] =='7+') 
    & (dia['Glucose'] >= 125) &  (dia['BP_Cat'] == 'Hoch')
).astype(int)
```

```python
dia_encoded['Risiko_ins_preg'] = (
    (dia['Insulin'] >= 166) & (dia['Preg_Cat'] =='7+') 
    &  (dia['BP_Cat'] == 'Normal')
).astype(int)
```

```python
dia_encoded['Risk_Age_Glu'] = (
    (dia['Age'] >= 36) & 
    (dia['Glucose'] >= 100)
).astype(int)
```

```python
dia_encoded['Risk_Ins_BMI'] = (
    (dia['Insulin'] >= 166) & 
    (dia['BMI'] >= 25)
).astype(int)
```

```python
dia_encoded['Risk_HighInsulin_Diabet'] = (
    (dia['Glu_Cat'] == 'Diabet') & 
    (dia['Insulin'] >= 166)
).astype(int)
```

```python
dia_encoded['Risk_Prediabet'] = (
    (dia['BP_Cat'] == 'Hoch') & (dia['ST_Cat'] == 'Dick') &
    (dia['Glu_Cat'] == 'Prediabet')
).astype(int)
```

```python
X = dia_encoded.drop(['Outcome', 'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Age', 'DiabetesPedigreeFunction'], axis=1)
y = dia_encoded['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=17)

rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:\n", classification_report(y_test, y_pred))

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
